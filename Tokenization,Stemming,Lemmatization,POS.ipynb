**SENTENCE TOKENIZATION**

```python
import nltk
nltk.download('punkt_tab') # Download the necessary tokenization models
from nltk.tokenize import sent_tokenize
```

```python
def tokenize_sentences(text):
    sentences = sent_tokenize(text)
    return sentences
```

```python
text = "I am Learning Natural Language Processing. It is for classification, tokenization, stemming, tagging, parsing, etc."
#print(tokenize_sentences(text))
```

```python
# Print tokenized sentences
sentences = tokenize_sentences(text)
for i, sentence in enumerate(sentences):
  print(f"Sentence {i+1}: {sentence}")
```

**WORD TOKENIZATION**

```python
from nltk.tokenize import word_tokenize
def tokenize_words(text):
  words = word_tokenize(text)
  return words
```

```python
# Example text
text = "I am Learning Natural Language Processing."
# Tokenize words
words = tokenize_words(text)
# Print tokenized words
print(words)
```

**LEMMATIZATION**

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Download required NLTK data (run once)
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
str_lem="the friends of love scenes"
words=word_tokenize(str_lem)
```

```python
lemmatizer = WordNetLemmatizer()
lem_words=[lemmatizer.lemmatize(word) for word in words]
print(lem_words)
```

**STEMMING**

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
str_stem="welcome to discovery of animals planet."
```

```python
words=word_tokenize(str_stem)
print(words)
```

```python
stem_words=[stemmer.stem(word) for word in words]
print(stem_words)
```
